```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data2.csv", index_col=0)
```

```python
df.head()
```

```python
from sklearn.feature_selection import SelectKBest, chi2

X = df.drop('Diagnosis', axis=1)
y = df['Diagnosis']

Feature_Names = X.columns

# Calculate the chi-square scores between each feature and the target variable
k=200
selector = SelectKBest(score_func=chi2, k=k)
X_new = selector.fit_transform(X, y)

# Print the selected features
selected_feature_indices = selector.get_support(indices=True)
selected_features = Feature_Names[selected_feature_indices]
print("Selected Features:", selected_features)

new_df = pd.DataFrame(X_new, columns=selected_features, index=df.index)
new_df['Diagnosis'] = y.values

new_df.head(25)
```

```python
new_df.to_csv("data5.csv")
```
