```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data2.csv", index_col=0)
```

```python
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

X = df.drop('Diagnosis', axis=1)
y = df['Diagnosis']

Feature_Names = X.columns


# Set up the Random Forest Classifier
clf = RandomForestClassifier()
clf.fit(X, y)

# Get feature importances
importances = clf.feature_importances_

# Sort features based on their importance scores
sorted_indices = np.argsort(importances)[::-1]

# Select the top-k features with the highest importance scores
k=200
selected_features = Feature_Names[sorted_indices[:k]].tolist()

# Create a new DataFrame with the selected features and target variable
new_df = X[selected_features].copy()
new_df['Diagnosis'] = y.values

# Print the new DataFrame
print(new_df.head())
```

```python
new_df.to_csv("data6.csv")
```
