```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset into a pandas DataFrame
# Replace 'your_dataset.csv' with the actual path to your dataset
data = pd.read_csv(
    'C:/Users/ACER/OneDrive - University of Jaffna/UOJ/Education/Research/Data Sets/GSE140842/Feature Selection\Information Gain\data_k_12.csv')

# Assuming your dataset has 13000 features, let's say we want to visualize the first 10 features
# You can adjust the range as needed
num_features_to_plot = 12

# Create a new figure with subplots
fig, axes = plt.subplots(3, 4, figsize=(20, 12))

# Flatten the axes array to make it easier to iterate through
axes = axes.flatten()

# Loop through the first num_features_to_plot features and create distribution plots
for i in range(num_features_to_plot):
    # Create a histogram with KDE
    sns.histplot(data=data, x=data.columns[i+1], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {data.columns[i+1]}')  # Set the title
    axes[i].set_xlabel(data.columns[i+1])  # Set the x-axis label
    axes[i].set_ylabel('Frequency')  # Set the y-axis label

# Adjust the layout of subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset into a pandas DataFrame
# Replace 'your_dataset.csv' with the actual path to your dataset
data = pd.read_csv(
    'C:/Users/ACER/OneDrive - University of Jaffna/UOJ/Education/Research/Data Sets/GSE140842/Feature Selection/Information Gain/data_k_12.csv')

# Assuming your dataset has 13000 features, let's say we want to visualize the first 12 features
# You can adjust the range as needed
num_features_to_plot = 12

# Create a new figure with subplots
fig, axes = plt.subplots(3, 4, figsize=(20, 12))

# Flatten the axes array to make it easier to iterate through
axes = axes.flatten()

# Loop through the first num_features_to_plot features and create scatter plots or box plots
for i in range(num_features_to_plot):
    # Check if the feature is numerical (you can adjust the condition based on your data)
    if data[data.columns[i+1]].dtype in ['int64', 'float64']:
        # Create a scatter plot
        sns.scatterplot(
            data=data, x=data.columns[i+1], y=data.columns[13], ax=axes[i])
        axes[i].set_title(
            f'Relationship between {data.columns[i+1]} and Target')
        axes[i].set_xlabel(data.columns[i+1])
        axes[i].set_ylabel(data.columns[13])
    else:
        # Create a box plot for categorical features
        sns.boxplot(
            data=data, x=data.columns[13], y=data.columns[i+1], ax=axes[i])
        axes[i].set_title(f'Box Plot of {data.columns[i+1]} by Target')
        axes[i].set_xlabel(data.columns[13])
        axes[i].set_ylabel(data.columns[i+1])

# Adjust the layout of subplots
plt.tight_layout()

# Show the plot
plt.show()
```
