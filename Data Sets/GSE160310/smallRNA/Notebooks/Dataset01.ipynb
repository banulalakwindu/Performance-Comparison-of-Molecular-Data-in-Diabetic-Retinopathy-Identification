```python
import pandas as pd

# Specify the file path to your text file
file_path = r'C:\Users\ACER\OneDrive - University of Jaffna\UOJ\Education\Research\Data Sets\GSE160310\smallRNA\GSE160308_human_retina_DR_smallRNA_counts.txt'

# Initialize empty lists to store data
data = []

# Read the text file line by line
with open(file_path, 'r') as file:
    for line in file:
        # Split each line into values using a comma as the delimiter
        values = line.strip().split(',')
        data.append(values)

# Create a DataFrame from the extracted data
df = pd.DataFrame(data)

# # Set the first row as the column headers
df.columns = df.iloc[0]

# # Drop the first row, which contains the column names
df = df[1:]

# Reset the index
df.reset_index(drop=True, inplace=True)

# Display the DataFrame
df
print(df)

df.to_csv('data0.csv', index=False)
```
