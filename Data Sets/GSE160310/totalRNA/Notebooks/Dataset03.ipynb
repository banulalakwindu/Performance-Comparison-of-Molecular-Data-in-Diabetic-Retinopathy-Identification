```python
import pandas as pd
import numpy as np
```

```python
df_features = pd.read_csv("data1.csv", index_col=0)
df_targets = pd.read_csv("Dignosis.csv", index_col=0)
```

```python
df_features.head()
```

```python
df_targets.head()
```

```python
X = df_features
y = df_targets['Disese Group']

Feature_Names = X.columns

# Calculate the Pearson correlation coefficient between each feature and the target variable
correlations = X.apply(lambda feature: np.abs(feature.corr(y)))

# Sort the features based on their absolute correlation values
sorted_features = correlations.sort_values(ascending=False)

# Select the top k features with the highest absolute correlation
k = 200  # Number of top features to select
selected_features = sorted_features[:k].index.tolist()

# Create a new DataFrame with the selected features and target variable
new_df = X[selected_features].copy()
new_df.loc[:, 'Disese Group'] = y.values

# Print the new DataFrame
print(new_df.head())
```

```python
new_df.to_csv("data3.csv")
```
